# Tabela dinâmica (Tratamento carteira de ações)
 
 
 __Nos permite remodelar a tabela de acordo com a necessidade vigente__
 
 * pivot_table(values=None, index=None, columns=None, aggfunc='mean', fill_value=None, margins=False, dropna=True, margins_name='All', observed=False)

```python
import pandas as pd
import numpy as np
```

```python
file = 'acoes_ficticias.csv'
df = pd.read_csv(file,delimiter=';') # delimiter muda o padrão dos separadores ( em pandas o padrão é a virgula)
df
```

```python
df.head()
```

```python
df.columns
```

```python
#Agrupando todos os papeis do mesmo grupo e somando suas quantidades
# aggfunc = '' default mean
pd.pivot_table(df,index='Papel',values='Quantidade',aggfunc='sum')
```

```python
# Adicionando uma nova coluna no data frame

df['valor_total'] = df.preco_compra*df.Quantidade

df.head(5)
```

```python
soma_qtd_valor = pd.pivot_table(df,index='Papel', values=['Quantidade','valor_total'], aggfunc='sum')

soma_qtd_valor
```

```python
soma_qtd_valor['pmedio'] = soma_qtd_valor.valor_total/soma_qtd_valor.Quantidade
soma_qtd_valor
```
