## Criando um DataFrame a partir de um dicionário

No exemplo abaixo temos um data frame composto por um dicionário, que por sua vez, é composto por um atributo chave - valor. O valor é uma serie pandas.

No contexto de um DataFrame as chaves servem como nome das colunas, e as series são os dados de dada coluna.

* pd.DataFrame({'key_1':pd.Series(...), 'key_n':pd.Series(...)})



```python
dic_2 ={'coluna 1': np.arange(10)*5, 'coluna 2': np.arange(10)**2, 'coluna 3' : np.ones (10)}

dic_2
```

```python
dic_2['coluna 1']
```

Criando um DataFrame a Partir de uma coluna

```python
df1 = pd.DataFrame(dic_2)
df1
```

```python
type(df1)
```

```python
df1['coluna 1'] # Se usarmos type em df1['coluna 1']  veremos que se trata de uma serie
```
