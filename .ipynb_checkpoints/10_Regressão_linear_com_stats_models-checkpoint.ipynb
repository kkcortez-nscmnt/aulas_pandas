# Regressão linear com stats models

Estrutura:

Sejam x e y dois arrays, series ou listas, onde:

*__Y = cte + cX__*


| ação | comentário|
|:-|:-|
| import **statsmodels.api** as **sm**| importa a biblioteca|
| X = sm.add_constant(X)| adiciona a constante da regressão linear, se desejável|
| est = sm.OLS(Y,X).fit()| gera o objeto de regressão linear baseada em Xe Y|
| est.params|retorna os parametros da regressão linear|
| print(est.summary())| imprime um relatório estatístico da regressão linear|

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
df = pd.DataFrame({'x':[0, 1, 2, 3, 4, 5,],
                  'y':[2, 3, 5, 10, 15, 20]})
df
```

```python
plt.figure()
plt.suptitle("Regressão")

plt.plot(df['x'], df['y'], 'ro')
plt.ylabel('Y')
plt.xlabel('X')

plt.grid()
plt.show()
```

```python
# regressão linear: encontrar uma constante tal que t = ax + c.
# métodos dos minimos quadrados pela biblioteca stats models

X = df['x'] # cria uma serie com os dados x
Y = df['y']
```

```python
est = sm.OLS(Y,X).fit() # cria um objeto de regressão linear
```
