# Propiedade DataFrame

DataFrames e Series em pandas compartilham propiedades e métodos com arrays numpy.

exemplo:

shape, size, dtypes, columns, index( retorna linhas), values(retona array)



|Métodos aplicados ao DataFrame| Descrição |
|:-|:-|
|.head(n)| exibe as n primeiras linhas|
|.tail(n)| exibe as n ultimas linhas|
|.to_csv(path/name)| exporta o DF para um arquivo csv (esse método existe para vários outros formatos)|
|.min(axis), .max(axis)| similar aos métodos numpy|
|.cumsum(), mean()| similar aos métodos numpy|
|.value_counts()| retorna a frequencia de determinado valor(aplicar para séries)|
|.sort_values(by=column) |ordena valores|

```python
import pandas as pd
import numpy as np
```

## Propiedades DataFrame

```python
lista_a =['a', 'b', 'c', 'd', 'a', 'd', 'c']
lista_b = list(range(7))
dic = {'string': lista_a, 'numeros': lista_b}
dic
```

```python
df = pd.DataFrame(dic)
df
```

```python
print(df.size)
print(df.ndim)
print(df.shape)
df.dtypes
```

```python
df.columns
```

```python
list(df.columns)
```

```python
df.index
```

### Podemos criar um data frame a partir de um array ou um array a partir de um data frame

```python
print(df.values)
print(df.values.shape)
print(type(df.values))
```
