# Visualização Gráfica com  Pandas

O pandas possui em sua composição objetos de vizualização inerentes da biblioteca matplotlib, dos quais, podemos acessalos sem necessariamente importar a bilioteca.y

* df.plot(x='x', y='y')

* df.plot.bar(x='x', y='y')

* df.plot.pie(subplots=True ou y ='y')

* df.plot.scatter(x='x', y='y')

https://pandas.pydata.org/docs/user_guide/visualization.html#visualization-hist

# Plotando um Gráfico Simples - Series

```python
import pandas as pd
import numpy as np

arr = np.arange(10)
arr

serie_01 = pd.Series(arr.cumsum())
serie_01
```

### __Lembrete:__

Séries são estruturas de dados unidimensionais indice/valor.

Indices não são colunas. Uma série é um objeto unidimensional possuindo somente uma coluna x n linhas

```python
serie_01.plot()

# os indices entram nos valores das abscissas
# os valores entra no eixo das ordenadas
```

# Plotando um gráfico elaborado - dataframe

```python
df1= pd.DataFrame({'a':np.arange(50)**2,
                   'b':np.arange(50)*3,
                   'c':np.arange(50)*5,})

df1.head(5)
```

```python
df1.plot()
```

```python
df1.plot(x ='c', y='a')
```

```python
df1.plot(x='c', y=['a','b'])
```

# Gráfico de barras

```python
file = 'acoes_ficticias.csv'
df = pd.read_csv(file,delimiter=';')
df

pd.pivot_table(df, index='Papel',values='Quantidade',aggfunc='sum')
```

```python
df['valor_total'] = df.preco_compra*df.Quantidade
df.head(5)
```

```python
soma_qtd_valor = pd.pivot_table(df,index="Papel", values=['Quantidade', 'valor_total'], aggfunc='sum')
soma_qtd_valor
```

```python
soma_qtd_valor.plot.bar(y='Quantidade')
```

# Gráfico de pizza

```python
soma_qtd_valor.plot.pie(subplots=True)
```

```python
soma_qtd_valor.plot.pie(y='Quantidade')
```
